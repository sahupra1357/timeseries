1. ***Q:*** What is sleep apnea?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: Sleep apnea refers to the condition where an individual's breathing frequently stops and starts throughout the night.
2. ***Q:*** What is the physiological mechanism of sleep apnea?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: There are two main types of sleep apnea, both result in frequent cessation of breathing during the night. The most common type is obstructive sleep apnea (OSA). OSA is caused by blockage of the airway when the soft tissue in the back of the throat collapses during sleep. The second type is central sleep apnea (CSA). CSA occurs when the brain fails to tell the lungs to breathe during sleep. As this signal is lost, the lungs do not take in the air that the body needs.
3. ***Q:*** What is the current prevalence of sleep apnea in the U.S?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: Sleep apnea affects approximately 26% of US adults, of whom about 90% are undiagnosed. 
4. ***Q:*** How is sleep apnea diagnosed?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: Based on patient symptoms and sleep history (reported by either the patient or by the person who shares the patient's bed or household) they may be referred to a sleep disorders clinic. The patient will typically undergo a polysomnography (PSG) or a home sleep apnea test (HSAT) which monitors the patient overnight, recording various types of physiological signals such as electroencephalogram (EEG; monitors brain function), nasal flow (monitors airflow during sleep), abdomen and/or chest effort (monitors the contraction and expansion of the abdomen or chest), blood oxygen level (monitors the oxygen levels in the blood), electromyography (EMG; monitors leg movements), and potentially other signals as well. The full night recording is then typically reviewed by a Registered Polysomnographic Technologist (RPSGT), assessing the total amount of hours the patient slept, as well as identifying all of the instances where the patient either stopped breathing completely (apnea) or had a decrease in airflow (hypopnea) during the night. Based on the results of this test, the physician is then able to diagnose the patient with having sleep apnea and recommends the next steps for treatment.
5. ***Q:*** What is the Apnea-Hypopnea Index (AHI) and how is it used to diagnose sleep apnea severity? How many sleep apnea severity levels are there?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: The AHI is calculated by counting the number of instances the patient had apneas and hypopneas, and dividing that number by the total amount of hours the patient slept during the night. The AHI is used to generate a diagnosis for the patient. If the patient has an AHI of less than 5 events/hour, the patient receives a negative diagnosis. If the patient has an AHI of 5-15 events/hour, the patient is diagnosed with mild sleep apnea. If the patient has an AHI of 15-30 events/hour, the patient is diagnosed with moderate sleep apnea. If the patient has an AHI greater than 30, the patient is diagnosed with severe sleep apnea. Based on the severity of the diagnosis, the physician can recommend different types of sleep apnea treatments.
6. ***Q:*** What are potential treatments for sleep apnea?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: One option is Continous Positive Airway Pressure (CPAP). CPAP is used to treat obstructive sleep apnea by transmitting pressurized air through the nose, mouth, or both, which applies pressure on the airway to reduce the likelihood the airway will collapse and obstruct breathing. Another treatment option is hypoglossal nerve stimulation (HGNS). HGNS is an implantable device that connects via an electrode to the hypoglossal nerve and utilizing electrical stimulation, stimulates the hypoglossal nerve which relieves upper airway obstruction by triggering tongue movements. Other treatments include various mouth and nose appliances, structural mouth and throat surgery, as well as life changes such as weight loss.
8. ***Q:*** How does medicare define compliance for a CPAP machine?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: A patient is compliant with a CPAP machine if they use the machine for more than 4 hours for at least 70% of the days over the course of 90 days. 
9. ***Q:*** What percentage of patients are compliant with a CPAP machine?<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A: Studies show that somewhere between 29% and 83% of patients do not meet the criteria for compliance due to removing the CPAP early in the night and/or skipping use altogether. Thus, being able to forecast when a patient is about to become non-compliant with the CPAP machine may allow clinicians to identify struggling patients. The clinicians can then reach out to the patients and support them in such a way that will increase their likelihood of becoming compliant with the CPAP machine.

Note: You should define `window=30` and `look=30`.

```python
def preprocess_data(df, window, look):
  """ Restructures the data for the forecasting tasks.

  Args:
    df (pd.DataFrame): A dataframe containing the number of minutes of CPAP usage per day
    window (int): An integer specifing the number of days to use for forecasting
    look (int): An integer specifing the number of days ahead the model should forecast

  Returns:
    x (np.array): A numpy array containing all examples to use for forecasting. 
              Each example represents an array of length `window` that 
              containes the number of CPAP usage days to use for forecasting.
    y (np.array): A numpy array containing all labels for forecasting. Each example represents
        an arry of length `look` that contains the number of CPAP usage days to forcast.
  """
  #### TODO: Implement the function here
  df = df.filter(regex='Day', axis=1)
  num_days = len(df.columns)-6
  nub_user = df.shape[0]
  x = np.zeros((nub_user,num_days - window - look, window))
  y = np.zeros((nub_user,num_days - window -look, look))
  for i in range(df.shape[0]):
    df_temp = train_data.iloc[i][9:]
    in_start = 0
    for j in range(x.shape[1]):
      in_end = in_start + window
      out_end = in_end + look
      if out_end <= num_days:
        x[i,j,:] = np.array([df_temp[in_start:in_end]])
        y[i,j,:] = np.array([df_temp[in_end:out_end]])
      in_start += 1
  return x, y
```

### Define feature and label transformations:
Below you can (but you don't have to) define two functions, one for the features and one for the labels. The functions should apply any transformations that you think would be beneficial for the forecasting tasks.

***Tip: Since it will probably be difficult for the models to forecast raw hours or minutes of CPAP usage for all 30 days based on the previous 30 days, we recommend you think of a way to compress the next 30 days of CPAP usage into an easier output for the model. For example, one option is to compress the next 30 days of CPAP usage into 3 values, one value represents the number of days the patient used the CPAP for 0 hours, the second value represents the number of days the patient used the CPAP for < 4 hours, and the third value represents the number of days the patient used the CPAP for > 4 hours.***

***Note: The only thing you should not change about the features and labels is that the features should always represent `window` days of CPAP usage data and the labels should always represent the next `look` days of CPAP usage data.***


```python
def transform_features(x):
  """ Apply feature transformations.

  Args:
    x (np.array): The feature data.

  Returns:
    x_transformed (np.array): Transformed features
  """
  x_transformed = np.zeros((x.shape[0],x.shape[1], x.shape[2], 4))
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      for k in range(x.shape[2]):
        x_transformed[i,j,k,0]=np.float16(x[i,j,k])
        if x[i,j,k] == 0:
          x_transformed[i,j,k,1]=np.float16(1)
        else:
          x_transformed[i,j,k,1]=np.float16(0)
        
        if x[i,j,k] >=0 and x[i,j,k] <= 240:
          x_transformed[i,j,k,2]=np.float16(1)
        else:
          x_transformed[i,j,k,2]=np.float16(0)

        if x[i,j,k] > 240:
          x_transformed[i,j,k,3]=np.float16(1)
        else:
          x_transformed[i,j,k,3]=np.float16(0)
  return x_transformed

def transform_labels(y):
  """ Apply label transformations.

  Args:
    y (np.array): The label data.

  Returns:
    y_transformed (np.array): Transformed labels
  """
  y_transformed = np.zeros((y.shape[0],y.shape[1], y.shape[2], 4))
  for i in range(y.shape[0]):
    for j in range(y.shape[1]):
      for k in range(y.shape[2]):
        y_transformed[i,j,k,0]=y[i,j,k]
        if y[i,j,k] == 0:
          y_transformed[i,j,k,1]=1
        else:
          y_transformed[i,j,k,1]=0
        
        if y[i,j,k] >0 and y[i,j,k] <= 240:
          y_transformed[i,j,k,2]=1
        else:
          y_transformed[i,j,k,2]=0

        if y[i,j,k] > 240:
          y_transformed[i,j,k,3]=1
        else:
          y_transformed[i,j,k,3]=0
  return y_transformed
```

```python
def target_function(p, x):
    m, c = p
    return m*x + c

def demingRegressor(x,y):
  #  model fitting.
  odr_model = odr.Model(target_function)
 
  # Create a Data object using sample data created.
  data = odr.Data(x, y)
 
  # Set ODR with the model and data.
  ordinal_distance_reg = odr.ODR(data, odr_model,beta0=[0.2, 1.])
 
  # Run the regression.
  output = ordinal_distance_reg.run()
  return output

```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def evaluate_forecasts(y_true, y_pred):
  scores = list()
  # calculate an RMSE score for each day
  for i in range(y_true.shape[1]):
    # calculate mse
    mse = mean_squared_error(y_true[:, i], y_pred[:, i])
    # calculate rmse
    rmse = sqrt(mse)
    # store
    scores.append(rmse)
  # calculate overall RMSE
  s = 0
  for row in range(y_true.shape[0]):
    for col in range(y_true.shape[1]):
      s += (y_true[row, col] - y_pred[row, col])**2
  score = sqrt(s / (y_true.shape[0] * y_true.shape[1]))
  return score, scores
```

```python
def get_ml_model(module_name, model_class, **hparams):
  """ Returns an intialized ML model.


  Args:
    module_name (str): The name of the module to use, e.g. `sklearn.linear_model`
    module_class (str): The name of the model class to use, e.g. `LogisticRegression`
    **hparams (dict): a dictionary containing all necessary hparams for the chosen model
  
  Returns:
    model : An initialized model
  """
  parent_module = importlib.import_module(module_name)
  model_class = eval('parent_module.' + str(model_class))
  hyper_params = {'max_depth': 9, 
                'gamma': 5, 
                'subsample': 0.7, 
                'colsample_bytree':0.8, 
                'learning_rate': 0.3,
                'max_bin' : 512
               }
  model = model_class(hyper_params)
  return model
```

```python
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from numpy import array
```

```python
def train_ml_model(model, x_train, y_train):
  """ Returns a trained ML model.

  Args:
    model : An initialized model
    x_train (np.array): A numpy array containing the input features
    y_train (np.array): A numpy array containing the output labels
  
  Returns:
    model : A trained model
  """
  ts = time.time()
  xgb = XGBRegressor(objective='reg:squarederror', n_estimators=1000)
  steps = list()
  steps.append(('standardize', StandardScaler()))
  steps.append(('normalize', MinMaxScaler()))
  steps.append(('model', xgb))
  
  # create pipeline
  pipeline = Pipeline(steps=steps)
  pipeline.fit(x_train.reshape(x_train.shape[0]*x_train.shape[1],x_train.shape[2]),y_train.reshape(y_train.shape[0]*y_train.shape[1],y_train.shape[2]))
  print('Training took: {0}s'.format(time.time()-ts))
  return pipeline
```

```python
# make a recursive multi-step forecast
def forecast(model, input_x, n_input):
  yhat_sequence = list()
  input_data = [x for x in input_x]
  #print(input_data)
  #input_data = input_x
  for j in range(30):
		# prepare the input data
    X = array(input_data[-n_input:]).reshape(1, n_input)
		# make a one-step forecast
    yhat = model.predict(X)[0]
		# add to the result
    yhat_sequence.append(yhat)
		# add the prediction to the input
    input_data.append(yhat)
  return yhat_sequence
```

```python
def get_dl_model(input_shape, optimizer, lr, loss, strategy):
  """ Returns a compiled DL keras model.


  Args:
    input_shape (tuple): The shape of the input
    optimizer (str): The name of the keras supported optimizer to use for training
    lr (float): Learning rate to use for training
    loss (str): The name of keras supported loss to use for training
  
  Returns:
    model (tf.keras.Functional): A compiled keras model
  """
  #with strategy.scope():
  inputs = tf.keras.Input(batch_shape=input_shape, dtype=tf.float32)
  #### TODO: Implement the rest of the model here

  n_steps_in = input_shape[0]
  n_features = input_shape[1]
  n_steps_out = 30
  # define model
  model = Sequential()
  model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps_in, n_features)))
  model.add(LSTM(50, activation='relu'))
  model.add(Dense(n_steps_out))
  model.add(Activation('linear'))    
  model.summary() # Don't remove this
  optimizer_obj = tf.keras.optimizers.get({'class_name': optimizer, 'config': {'learning_rate': lr}}) # Don't remove this
  model.compile(optimizer=optimizer_obj, loss=loss) # Don't remove this
  return model
```
